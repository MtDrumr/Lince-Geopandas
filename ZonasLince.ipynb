```python
lista_zonas1=[]
lista_zonas2=[]
lista_var=1

for x in list(zona0.columns):
    zona1=zona0.loc[zona0[x].isna()==False]
    zona2=pd.DataFrame(zona1[x]).rename(columns={x:"npunto"})
    zona3=list(zona2.merge(data, how='left', on='npunto')["punto"])

    lat_point_list=[zona3[x][0] for x in range(len(zona3))]
    lon_point_list=[zona3[x][1] for x in range(len(zona3))]

    polygon_geom = Polygon(zip(lon_point_list, lat_point_list))
    crs = {'init': 'epsg:4326'}
    polygon = geopandas.GeoDataFrame(index=[0], crs=crs, geometry=[polygon_geom])     
    lista_zonas1.append("Zona {}".format(lista_var))
    lista_zonas2.append(polygon)
    lista_var+=1
    

lince5=pd.DataFrame(data={"ZONA1":lista_zonas1,"geometry":lista_zonas2}).copy()
lince5
```

```python
mapa1=folium.Map(location = [-12.085,-77.04], tiles='OpenStreetMap', zoom_start = 15)

for x in range(lince5.shape[0]):
    geo_j = folium.GeoJson(lince5["geometry"][x]["geometry"][0],
        style_function=lambda x: {"color":"black","weight": 3.5,"fillOpacity":0})
    folium.Popup(lince5["ZONA1"][x]).add_to(geo_j)
    geo_j.add_to(mapa1)

delay=5
fn='map1.html'
tmpurl='file://{path}/{mapfile}'.format(path=os.getcwd(),mapfile=fn)
mapa1.save(fn)

browser = webdriver.Chrome(ChromeDriverManager().install())
browser.get(tmpurl)
#Give the map tiles some time to load
time.sleep(delay)
browser.save_screenshot('map1.png')
browser.quit()
```
